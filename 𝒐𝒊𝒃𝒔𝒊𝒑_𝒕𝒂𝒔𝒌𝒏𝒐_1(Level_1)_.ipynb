# ***Time Series Analysis***

```python
# Creating a 'DATE' column from 'YEAR' and 'MONTH'
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(DAY=1))

# Grouping data by date and calculating total sales for each month
monthly_sales = df.groupby('DATE')[['RETAIL SALES', 'WAREHOUSE SALES']].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales['RETAIL SALES'], label='Retail Sales', color='blue', marker='o')
plt.plot(monthly_sales.index, monthly_sales['WAREHOUSE SALES'], label='Warehouse Sales', color='green', marker='o')

plt.title('Retail and Warehouse Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# ***Customer and Product Analysis Visualizations***

```python
# 1. Top Products by Sales (Retail and Warehouse)
top_products_retail = df.groupby('ITEM DESCRIPTION')['RETAIL SALES'].sum().sort_values(ascending=False).head(10)
top_products_warehouse = df.groupby('ITEM DESCRIPTION')['WAREHOUSE SALES'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_products_retail.plot(kind='bar', color='skyblue', label='Retail Sales')
plt.title('Top 10 Products by Retail Sales')
plt.ylabel('Sales Amount')
plt.xlabel('Item Description')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Top Suppliers by Sales
top_suppliers_retail = df.groupby('SUPPLIER')['RETAIL SALES'].sum().sort_values(ascending=False).head(10)
top_suppliers_warehouse = df.groupby('SUPPLIER')['WAREHOUSE SALES'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_suppliers_retail.plot(kind='bar', color='pink', label='Retail Sales')
plt.title('Top 10 Suppliers by Retail Sales',fontsize=15)
plt.ylabel('Sales Amount',fontsize=15)
plt.xlabel('Supplier',fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. Product Preference by Item Type
item_type_sales = df.groupby('ITEM TYPE')[['RETAIL SALES', 'WAREHOUSE SALES']].sum().sort_values(by='RETAIL SALES', ascending=False)

plt.figure(figsize=(10, 6))
item_type_sales['RETAIL SALES'].plot(kind='line', color='black', label='Retail Sales',marker='*')
plt.title('Sales by Item Type')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 4. Heatmap: Correlation between Retail and Warehouse Sales
sales_correlation = df[['RETAIL SALES', 'WAREHOUSE SALES', 'RETAIL TRANSFERS']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(sales_correlation, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Between Sales Metrics',fontsize=15)
plt.tight_layout()
plt.show()
```
